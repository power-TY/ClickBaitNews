```python
import math
import numpy as np
import logging
from datetime import datetime

import torch
from torch.utils.data import DataLoader
from datasets import load_dataset
from sentence_transformers import SentenceTransformer, models, LoggingHandler, losses, util
from sentence_transformers.readers import InputExample
```

```python
# logger
logger=logging.basicConfig(
    format="%(asctime)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
    level=logging.INFO,
    handlers=[LoggingHandler()],
)

pretrained_model_name = 'klue/roberta-base'
nli_num_epochs = 5
train_batch_size = 128

nli_model_save_path = '/content/drive/MyDrive/TM/training_nli_by_onlineloss'+pretrained_model_name.replace("/", "-")+'-'+datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
# sts_model_save_path = 'output/training_sts_continue_training-'+pretrained_model_name.replace("/", "-")+'-'+datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
```

```python


def make_nli_input_example(dataset):
    '''
    Transform to InputExample
    '''
    input_examples = []
    for i in range(len(dataset)):
        sentence1 = dataset.loc[i, 'clickbaitTitle']
        sentence2 = dataset.loc[i, 'content']
        label = dataset.loc[i, 'clickbaitClass']
        input_examples.append(InputExample(texts=[sentence1, sentence2], label=label))

    return input_examples
```

```python
nli_train = make_nli_input_example(df_train_sample)
nli_valid = make_nli_input_example(df_val_sample)
nli_test = make_nli_input_example(df_test)
```

```python
len(nli_train)
```

```python
from sentence_transformers.evaluation import BinaryClassificationEvaluator

# Train Dataloader
train_dataloader = DataLoader(
    nli_train,
    shuffle=True,
    batch_size=train_batch_size,
)

# Evaluator val
dev_evaluator = BinaryClassificationEvaluator.from_input_examples(
    nli_valid,
    name="nli-dev",
    show_progress_bar=True
)

# Evaluator test
test_evaluator = BinaryClassificationEvaluator.from_input_examples(
    nli_test,
    name="nli-test",
    show_progress_bar=True
)
```

```python
# Load Embedding Model
embedding_model = models.Transformer(
    model_name_or_path=pretrained_model_name,
    max_seq_length=128,
    do_lower_case=True
)

# Only use Mean Pooling -> Pooling all token embedding vectors of sentence.
pooling_model = models.Pooling(
    embedding_model.get_word_embedding_dimension(),
    pooling_mode_mean_tokens=True,
    pooling_mode_cls_token=False,
    pooling_mode_max_tokens=False,
)

model = SentenceTransformer(modules=[embedding_model, pooling_model])
```

```python
import gc
torch.cuda.empty_cache()
gc.collect()
```

```python
# Use SoftmaxLoss, because NLI is Multi-class Classification task.
train_loss = losses.OnlineContrastiveLoss(model=model)

# warmup steps
warmup_steps = math.ceil(len(nli_train) * nli_num_epochs / train_batch_size * 0.1) #10% of train data for warm-up
logging.info("Warmup-steps: {}".format(warmup_steps))

# Training
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    evaluator=dev_evaluator,
    epochs=nli_num_epochs,
    evaluation_steps=int(len(train_dataloader)*0.1),
    warmup_steps=warmup_steps,
    show_progress_bar=True,
    output_path=nli_model_save_path
)
```

```python
# test case 성능결과 0.96은 결과 중 제일 높은 값, save path로 저장되는 csv파일 따로 참조
test_evaluator(model, output_path=nli_model_save_path )
```
